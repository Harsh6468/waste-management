```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option("display.max_columns",100)
pd.set_option("display.max_rows",100)
```

```python
# Load raw dataset (not preprocessed, for EDA readability)
df = pd.read_csv("../data/raw/waste_data.csv")
```

```python
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (8, 5)
```

```python
# 1. Distribution of Recycling Rate
sns.histplot(df['Recycling Rate (%)'], kde=True, bins=20, color="red")
plt.title("Distribution of Recycling Rate (%)")
plt.show()
```

```python
# 2. Recycling Rate by Waste Type
sns.boxplot(x='Waste Type', y='Recycling Rate (%)', data=df, palette="Set1")
plt.title("Recycling Rate by Waste Type")
plt.xticks(rotation=45)
plt.show()
```

```python
# 3. Recycling Rate by Disposal Method
sns.boxplot(x='Disposal Method', y='Recycling Rate (%)', data=df, palette="Set1")
plt.title("Recycling Rate by Disposal Method")
plt.show()
```

```python
# 4. Correlation Heatmap (numerical only)
plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include='number').corr(), annot=True, cmap='YlGnBu', fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# 5. Yearly Trend of Recycling Rate
sns.lineplot(x='Year', y='Recycling Rate (%)', data=df, ci=None, marker="o", color="blue")
plt.title("Yearly Trend of Recycling Rate")
plt.show()
```

```python
# 6. Top 10 Cities by Avg Recycling Rate
city_avg = df.groupby("City/District")["Recycling Rate (%)"].mean().sort_values(ascending=False).head(10)
sns.barplot(x=city_avg.values, y=city_avg.index, palette="viridis")
plt.title("Top 10 Cities by Avg Recycling Rate")
plt.xlabel("Average Recycling Rate (%)")
plt.show()
```
